```python
import pandas as pd
import numpy as np

df1 = pd.read_csv("Heights+and+Weights.csv")
df1
```

```python
import matplotlib.pyplot as plt

plt.title("Heights and Weights")
plt.scatter(df1.Height,df1.Weight)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

```python
#value 轉換為ndarray
x = df1.iloc[:,:-1].values
y = df1.iloc[:,-1].values
```

```python
#導入model
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
#預測結果
model.predict([[1.73]])
```

```python
# y =mx+b
model.coef_
```

```python
#b
model.intercept_

#y = 59.9x -36.76
```

```python
plt.title("Height and Weight (Training)")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.scatter(x_train,y_train,color = "r")
plt.plot(x_train,model.predict(x_train),color="b")
plt.show()
```

```python
# 看測試集
plt.title("Height and Weight (Testing)")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.scatter(x_test,y_test,color = "r")
plt.plot(x_train,model.predict(x_train),color="b")
plt.show()
```
